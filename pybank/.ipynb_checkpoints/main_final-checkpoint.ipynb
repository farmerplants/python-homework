```python
#imports

import csv
from pathlib import Path

#set path for csv
csvpath = Path('/Users/', 'tylerhubbard', 'Downloads', 'budget_data.csv')

#initialize all variables to be used:

#total months in file (column 1)
tot_months = 0

#net total amount of PnL over total period (column 2)
net_pnl = 0

#average of the changes in PnL over total period:
#changes in PnL I think would be (PnL2-PnL1) in which 
#PnL1 is the previous amount/PnL2 is the current amount
change = 0
avg_change = 0
pnl1 = 0
pnl2 = 0

#greatest increase/decrease in profits (date and amount) over total period
pnl_max = 0
pnl_min = 0
pnl_max_month = ''
pnl_min_month = ''

#final analysis for print out
final_output = ''

#execute csv math
with open(csvpath, 'r') as csvf:
    csvr = csv.reader(csvf, delimiter=',')
    next(csvr)
    for row in csvr:
        #set value for current amount
        pnl2 = int(row[1])
        if pnl1 != 0:
            #set value for change and total change
            change = pnl2 - pnl1
            avg_change = avg_change + change
        #set value for total number of months and total PnL
        net_pnl = net_pnl + pnl2
        tot_months = tot_months + 1
        #set values for greatest/worst profit days
        if change > pnl_max:
            pnl_max = change
            pnl_max_month = row[0]
        elif change < pnl_min:
            pnl_min = change
            pnl_min_month = row[0]
        pnl1 = pnl2

#print out the final output
#include math for average output
final_output = (
    f'Financial Analysis\n'
    f'------------------\n'
    f'Total Months: {tot_months}\n'
    f'Total: ${net_pnl}\n'
    f'Average Change: ${round(avg_change / (tot_months - 1), 2)}\n'
    f'Greatest Increase in Profits: {pnl_max_month} (${pnl_max})\n'
    f'Greatest Decrease in Profits: {pnl_min_month} (${pnl_min})\n'
)
print(final_output)

#textfile export to same directory
final_output_path = Path('/Users', 'tylerhubbard', 'umnfintech', 'python-homework', 'pybank', 'final_output.txt')
with open(final_output_path, 'w') as file:
    file.write(final_output)
```
